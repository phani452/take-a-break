```python
import tkinter as tk        #importing libraries
import webbrowser           #importing libraries
from time import *          #importing libraries
import sys                  #importing libraries
fp=open(r'C:\Users\Phaneendra\Desktop\file.txt')       #database file of users             
l=[]
l5=[]
a=[0]
sk=[]
global b
b=0
global c
c=0
l1=['https://www.youtube.com/watch?v=weeI1G46q0o',
    'https://www.youtube.com/watch?v=ZBDSNy4Yn9Q',
    'https://www.youtube.com/watch?v=Bg8Yb9zGYyA']      #list of videos
for line in fp:
    l5.append(line.lstrip().rstrip())
print(l5)
root = tk.Tk()             
root.title('take a break')       #title of application
def video1():                    #function to open ml article
    sleep(3)
    string = strftime('%H:%M:%S%p') 
    l.append('time'+':'+string+':'+'https://en.wikipedia.org/wiki/Machine_learning')
    webbrowser.open('https://en.wikipedia.org/wiki/Machine_learning')
def video2():                   #function to open AI article
    sleep(3)
    string = strftime('%H:%M:%S%p') 
    l.append('time'+':'+string+':'+'https://en.wikipedia.org/wiki/Artificial_intelligence')
    webbrowser.open('https://en.wikipedia.org/wiki/Artificial_intelligence')
def video3():                   #function to open python article
    sleep(3)
    string = strftime('%H:%M:%S%p') 
    l.append('time'+':'+string+':'+'https://en.wikipedia.org/wiki/Python')
    webbrowser.open('https://en.wikipedia.org/wiki/Python')
def video4():
    sleep(3)
    string = strftime('%H:%M:%S%p') 
    l.append('time'+':'+string+':'+'https://www.youtube.com/watch?v=YysWbPzkgpo')
    webbrowser.open('https://www.youtube.com/watch?v=h2JH0vqDcYc')
def history():                   #function to display the history of the user
    r3=tk.Tk()
    r3.title('history')
    count=0
    for i in l:
        count=count+1
        label8 = tk.Label(r3,text=str(count)+':'+i)
        label8.pack()
    r3.mainloop()
def shuffle():                   #function to shuffle a playlist
    count=1
    a.append(count)
    b=len(a)
    print(b)
def skip():                      #function to skip videos in a playlist
    countskip=2
    sk.append(countskip)
    c=len(sk)
    return c
    
def playlist():                    #fuction to run playlist of videos
    if b%2!=0 or b==1:
        
            for i in l1:
                webbrowser.open(i)
                sleep(10)
                string = strftime('%H:%M:%S%p')
                l.append('time'+':'+string+':'+i)
                
    else:
        import random
        l2=l1.copy()
        random.shuffle(l2)
        for i in l2:
            webbrowser.open(i)
            sleep(10)
            string = strftime('%H:%M:%S%p')
            l.append('time'+':'+string+':'+i)
def playlist1():
    ph=skip()
    if ph==1:
        for i in range(1,3):
            webbrowser.open(l1[i])
    else:
        for i in range(2,3):
            webbrowser.open(l1[i])

label_1 = tk.Label(root, text="Username:")           #login details
label_2 = tk.Label(root, text="Password:")

entry_1 = tk.Entry(root)
entry_2 = tk.Entry(root,show='*')

label_1.grid(row=0)
label_2.grid(row=1)
entry_1.grid(row=0, column=1)
entry_2.grid(row=1, column=1)
def video7():                                               #function to check whether user is there in database                          
    if entry_1.get() + entry_2.get() in l5:
        r = tk.Tk() 
        r.title('take a break')
        def video6():
            r.destroy()
        def time():                                          #function to display time
            string = strftime('%H:%M %p') 
            lbl.config(text = string) 
            lbl.after(1000, time)
            
        def countdown(count):                                #function for timer
            # change text in label        
            label['text'] = count

            if count > 0:
                # call countdown again after 1000ms (1s)
                r.after(1000, countdown, count-1)
            else:
                r.destroy()
                r2=tk.Tk()
                lb4=tk.Label(r2,text="get back to work",font = ('calibri', 50, 'bold'))
                lb4.pack()
                r2.wm_iconbitmap(r'C:\Users\Phaneendra\Desktop\pu.ico')
                r2.mainloop()
        
        #creating layout for the application....    
        button = tk.Button(r, text='Machine Learning', width=25,background = 'light blue',foreground = 'white',command=video1) 
        button.pack(anchor='sw')
        button1 = tk.Button(r, text='Artificial Intelligence', width=25,background = 'light blue',foreground = 'white',command=video2)
        button1.pack(anchor='sw')
        button2 = tk.Button(r, text='Python', width=25,background = 'light blue',foreground = 'white',command=video3)
        button2.pack(anchor='sw')
        button3 = tk.Button(r, text='Start', width=25,background = 'light blue',foreground = 'white',command=video4)
        button3.pack(anchor='sw')
        button4 = tk.Button(r, text='stop', width=25,background = 'light blue',foreground = 'white',command=video6)
        button4.pack(side='bottom')
        button10 = tk.Button(r, text='history', width=25,background = 'light blue',foreground = 'white',command=history)
        button10.pack(anchor='sw')
        button11 = tk.Button(r, text='playlist', width=25,background = 'light blue',foreground = 'white',command=playlist)
        button11.pack(anchor='sw')
        button12 = tk.Button(r, text='shuffle', width=25,background = 'light blue',foreground = 'white',command=shuffle)
        button12.pack(anchor='sw')
        button13 = tk.Button(r, text='skip', width=25,background = 'light blue',foreground = 'white',command=playlist1)
        button13.pack(anchor='sw')
        def timer():
            time() 
        def phani():
            countdown(10)
        

        button7 = tk.Button(r, text='start clock', width=25,background = 'light blue',foreground = 'white',command=timer)
        button7.pack(anchor='sw')
        button8 = tk.Button(r, text='start timer', width=25,background = 'light blue',foreground = 'white',command=phani)
        button8.pack(anchor='sw')
        lb2=tk.Label(r,text='clock',font=('calibri',20,'bold'))
        lb2.pack(anchor='ne')
        lbl = tk.Label(r, font = ('calibri', 30, 'bold'), 
                    background = 'light green', 
                    foreground = 'white') 
        lbl.pack(anchor='ne') 
        lb3=tk.Label(r,text='timer',font=('calibri',20,'bold'))
        lb3.pack(anchor='ne')
        label = tk.Label(r,font = ('calibri', 30, 'bold'),background = 'light green',foreground = 'white')
        label.pack(anchor='ne')
        r.wm_iconbitmap(r'C:\Users\Phaneendra\Desktop\pu.ico')
        r.mainloop()
    else:
        r1=tk.Tk()
        label1=tk.Label(r1,text='You are not an user!',font = ('calibri', 30, 'bold'))
        label1.pack()
        r1.mainloop()
        
        
button6=tk.Button(root,text='login',width=25,command=video7)
button6.grid(row=2,column=1)

checkbox = tk.Checkbutton(root, text="Keep me logged in")
checkbox.grid(columnspan=2)
s=strftime('%H')
if int(s)>12 and int(s)<17:
    label5 = tk.Label(root,text='good afternoon',font = ('calibri', 30, 'bold'),background = 'light green',foreground = 'white')
    label5.grid(columnspan=3)
elif int(s)<12:
    label6 = tk.Label(root,text='good morning',font = ('calibri', 30, 'bold'),background = 'light green',foreground = 'white')
    label6.grid(columnspan=3)
elif int(s)>17:
    label7 = tk.Label(root,text='good evening',font = ('calibri', 30, 'bold'),background = 'light green',foreground = 'white')
    label7.grid(columnspan=3)
else:
    label8 = tk.Label(root,text='good night',font = ('calibri', 30, 'bold'),background = 'light green',foreground = 'white')
    label8.grid(columnspan=3)
    
root.wm_iconbitmap(r'C:\Users\Phaneendra\Desktop\pu.ico') #icon
root.mainloop()
```
