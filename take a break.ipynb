```python
import tkinter as tk
import webbrowser 
from time import *
import sys

r1 = tk.Tk() 
r1.title('Counting')
def video1():
    sleep(3)
    webbrowser.open('https://www.youtube.com/watch?v=8EJ3zbKTWQ8')
def video2():
    sleep(3)
    webbrowser.open('https://www.youtube.com/watch?v=XejVB_fba04')
def video3():
    sleep(3)
    webbrowser.open('https://www.youtube.com/watch?v=YysWbPzkgpo')
def video4():
    sleep(3)
    webbrowser.open('https://www.youtube.com/watch?v=h2JH0vqDcYc')
def video7():
    r1.destroy()
def video5():
    r = tk.Tk() 
    r.title('Counting Seconds')
    def video6():
        r.destroy()
    def time(): 
        string = strftime('%H:%M %p') 
        lbl.config(text = string) 
        lbl.after(1000, time)
            
    def countdown(count):
        # change text in label        
        label['text'] = count

        if count > 0:
            # call countdown again after 1000ms (1s)
            r.after(1000, countdown, count-1)
        else:
            r.destroy()
            r2=tk.Tk()
            lb4=tk.Label(r2,text="get back to work",font = ('calibri', 50, 'bold'))
            lb4.pack()
            r2.mainloop()
        
            
    button = tk.Button(r, text='Start', width=25,background = 'red',foreground = 'white',command=video1)
    button.pack(anchor='sw')
    button1 = tk.Button(r, text='Start', width=25,background = 'red',foreground = 'white',command=video2)
    button1.pack(anchor='sw')
    button2 = tk.Button(r, text='Start', width=25,background = 'red',foreground = 'white',command=video3)
    button2.pack(anchor='sw')
    button3 = tk.Button(r, text='Start', width=25,background = 'red',foreground = 'white',command=video4)
    button3.pack(anchor='sw')
    button4 = tk.Button(r, text='stop', width=25,background = 'red',foreground = 'white',command=video6)
    button4.pack(side='bottom')
    def timer():
        time() 
    def phani():
        countdown(10)
        

    button7 = tk.Button(r, text='start clock', width=25,background = 'red',foreground = 'white',command=timer)
    button7.pack(anchor='sw')
    button8 = tk.Button(r, text='start timer', width=25,background = 'red',foreground = 'white',command=phani)
    button8.pack(anchor='sw')
    lb2=tk.Label(r,text='clock',font=('calibri',20,'bold'))
    lb2.pack(anchor='ne')
    lbl = tk.Label(r, font = ('calibri', 30, 'bold'), 
                background = 'purple', 
                foreground = 'white') 
    lbl.pack(anchor='ne') 
    lb3=tk.Label(r,text='timer',font=('calibri',20,'bold'))
    lb3.pack(anchor='ne')
    label = tk.Label(r,font = ('calibri', 30, 'bold'),background = 'purple',foreground = 'white')
    label.pack(anchor='ne')
    

    #button4.pack(side='right')
    r.mainloop()
button5 = tk.Button(r1, text='Start', width=25,command=video5)
button5.pack()
button6=tk.Button(r1,text='exit',width=25,command=video7)
button6.pack(side='bottom')
s=strftime('%H')
if int(s)>12 and int(s)<17:
    label5 = tk.Label(r1,text='good afternoon',font = ('calibri', 30, 'bold'),background = 'purple',foreground = 'white')
    label5.pack()
elif int(s)<12:
    label6 = tk.Label(r1,text='good morning',font = ('calibri', 30, 'bold'),background = 'purple',foreground = 'white')
    label6.pack()
elif int(s)>17:
    label7 = tk.Label(r1,text='good evening',font = ('calibri', 30, 'bold'),background = 'purple',foreground = 'white')
    label7.pack()
    
    
#button6.pack(side='right')
r1.mainloop()

```
